```python
"""
Getting the meta data in a list and the real words, pos tags, and lemmas in another list
"""
meta_data = []
stories = []
stop_chars = ['<', '>']
with open("January2016.txt") as infile:
        for line in infile:
            l = line.rstrip()
            if "<text" in l:
                meta_data.append(l)
            if l == "<story>":
                story = []
            elif l == "</story>":
                stories.append(story)
            elif not any(stop_char in l for stop_char in stop_chars):
                story.append(l)
#print(meta_data)
#stories[0]
```
